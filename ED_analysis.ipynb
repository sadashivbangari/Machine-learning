Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

Reading the data

```python
df= pd.read_csv('/content/automobile.csv')
```

displaying the 1st 5 rows of data set

```python
df.head()
```

Describe method

```python
df.describe()
```

Box plot

```python
sns.boxplot(x='num-of-cylinders',y='price',data=df)
```

scatter plot

```python
plt.scatter(df['engine-size'],df['price'])
plt.xlabel('engine-size')
plt.ylabel('price')
plt.show()
```

histogram

```python
count,bin_edges=np.histogram(df['peak-rpm'])
df['peak-rpm'].plot(kind='hist',xticks=bin_edges)
plt.xlabel('value of peak-rpm')
plt.ylabel('numbers of cars')
plt.grid()
plt.show()
```

Heat map

```python
sns.heatmap(df.isnull())
plt.show()
```

correlation

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```
